```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel("dataset.xlsx")

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(df['flow_duration'][:500], df['total_forward_packets'][:500], color='red')
plt.title("Line Plot (flow_duration vs total_forward_packets)")
plt.xlabel("flow_duration1")
plt.ylabel("total_forward_packets")

plt.subplot(2, 2, 2)
pie_values = df['total_forward_packets'][:5].abs()
plt.pie(pie_values, labels=df['flow_duration'][:5], autopct="%1.1f%%")
plt.title("Pie Chart (First 5 Values)")

plt.subplot(2, 2, 3)
plt.bar(df['flow_duration'][:10], df['total_forward_packets'][:10], color='red')
plt.title("Bar Chart (10 Samples)")
plt.xlabel("flow_duration")
plt.ylabel("total_forward_packets")


plt.subplot(2, 2, 4)
plt.scatter(df['flow_duration'][:500], df['total_forward_packets'][:500], color='red')
plt.title("Scatter Plot (flow_duration vs total_forward_packets)")
plt.xlabel("flow_duration")
plt.ylabel("total_forward_packets")

plt.tight_layout()
plt.savefig("dashboard.png",dpi=300,bbox_inches='tight')
plt.show()
```
